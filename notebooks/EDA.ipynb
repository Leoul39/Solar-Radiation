## Time Series Analysis


### Patterns by month
- The following plot function aggregates the solar radiation datasets by day. We'll also try to observe the rolling means of the column to see the distribution of the column.

```python
eda.plot_by_month(data=benin,column='GHI',name='benin')
```

```python
eda.plot_by_month(data=togo,column='DHI',name='togo')
```

### Patterns for a day
- The following plot function works for a specific day. It helps to analyze the distribution of values in certain columns throughout the day.

```python
eda.plot_by_day(data=benin,date='2021-08-19',name='benin')
```

```python
eda.plot_by_day(data=sierraleone,date='2022-02-09',name='sierraleone')
```

## Outlier Detection
- This outlier detector function works by using 2 different methods: `iqr` method and `z-score` method. Any column from any dataset can be entered and using the above methods, this function will output the amount of outliers.

```python
eda.detect_outliers(data=benin,column='WSgust',method='z-score',name='benin')
```

```python
eda.detect_outliers(data=togo,column='WSgust',method='z-score',name='togo')
```

```python
eda.detect_outliers(data=benin,column='WS',method='z-score',name='benin')
```
